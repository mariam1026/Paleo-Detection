# Paleo-Neutrino Detection Code

This notebook visually analyses SPIM data of the mica bioitite that has been processed through FIJI software to identify (count) particles. A particle could refer to a number of artefacts, spefically: ARTs, neutrinos or fission tracks. A slice is 187.25 x 187.25 µm or 1920 x 1920 pixels in size. The size for one pixel is approximately 0.1 µm².

```python
# Load CSV file
df = pd.read_csv(r'C:\Paleo-Detection\output-3\summary_all_slices2.csv')  # Raw string for Windows path

# Extract relevant columns
slices = df['Slice']
particle_counts = df['Particle Count']

# Compute average and standard deviation
mean = np.mean(particle_counts)
std = np.std(particle_counts)
print(f"The average particle count is: {mean:.2f}")

# Calculate sigma bounds, clipped at 0 for the lower bounds
plus_1sigma = mean + std
minus_1sigma = max(mean - std, 0)
plus_2sigma = mean + 2 * std
minus_2sigma = max(mean - 2 * std, 0)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(slices, particle_counts, linestyle='-', color='b', label='Particle Count')

# Shaded sigma regions
plt.fill_between(slices, minus_2sigma, plus_2sigma, color='red', alpha=0.2, label='±2σ Region')
plt.fill_between(slices, minus_1sigma, plus_1sigma, color='blue', alpha=0.3, label='±1σ Region')

# Mean line
plt.axhline(mean, color='black', linestyle='--', label=f'Mean = {mean:.2f}')

# Labels and layout
plt.xlabel('Slice')
plt.ylabel('Particle Count')
plt.title('Particle Count per Slice with σ Regions')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Below, the csv is read with no header (`header=none`), and assumes the first column contains the relevant slices - taking the maximum value in the first column as the 'particle count'. All the data in the slice file is used - no positional filtering. 

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Directory where CSVs are stored
data_dir = r'C:\Paleo-Detection\output_csv_summary_2-20'
max_slice = 3546

slice_numbers = []
particle_counts = []

for slice_num in range(1, max_slice + 1):
    file_path = os.path.join(data_dir, f'results_slice_{slice_num}.csv')
    if os.path.exists(file_path):
        try:
            df = pd.read_csv(file_path, header=0)  # Use header=0 to load with named columns

            # Check required columns exist
            if 'Y' not in df.columns:
                print(f"Slice {slice_num} is missing 'Y' column.")
                continue

            # Clean and filter
            df['Y'] = pd.to_numeric(df['Y'], errors='coerce')
            df = df.dropna(subset=['Y'])

            # Determine Y bounds for this slice
            y_min = df['Y'].min()
            y_max = df['Y'].max()

            y_lower_cutoff = y_min + 50
            y_upper_cutoff = y_max - 50

            # Apply Y-range cut
            df_cut = df[(df['Y'] >= y_lower_cutoff) & (df['Y'] <= y_upper_cutoff)]

            # Count remaining particles
            count = len(df_cut)
            slice_numbers.append(slice_num)
            particle_counts.append(count)

        except Exception as e:
            print(f"Error processing slice {slice_num}: {e}")
    else:
        print(f"Missing file: {file_path}")

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(slice_numbers, particle_counts, linestyle='-', color='blue')
plt.xlabel('Slice')
plt.ylabel('Particle Count (Y within 50 µm margin)')
plt.title('Filtered Particle Count vs Slice (excluding top/bottom 50 µm)')
plt.grid(True)

if particle_counts:
    plt.ylim(1, max(particle_counts))

plt.tight_layout()
plt.savefig('filtered_particle_count.png')
plt.show()
```

The graph below differs from the graph above as the CSVs are read with a header (`header=0`), and positional filtering is introduced - the top and bottom 50 µm of the slice is cut. The y-axis region stays fixed regardless of what is in each slice.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Directory where CSVs are stored
data_dir = r'C:\Paleo-Detection\output_csv_summary_2-20'
max_slice = 3546

# Define fixed hard Y-bounds (µm)
slice_top = 0
slice_bottom = 187.25
margin = 50

y_lower_cutoff = slice_top + margin          # 50 µm from top = 50
y_upper_cutoff = slice_bottom - margin       # 50 µm from bottom = 137.25

slice_numbers = []
particle_counts = []

for slice_num in range(1, max_slice + 1):
    file_path = os.path.join(data_dir, f'results_slice_{slice_num}.csv')
    if os.path.exists(file_path):
        try:
            df = pd.read_csv(file_path, header=0)

            if 'Y' not in df.columns:
                print(f"Slice {slice_num} missing 'Y' column.")
                continue

            df['Y'] = pd.to_numeric(df['Y'], errors='coerce')
            df = df.dropna(subset=['Y'])

            # Apply hard Y-bounds filter
            df_cut = df[(df['Y'] >= y_lower_cutoff) & (df['Y'] <= y_upper_cutoff)]

            count = len(df_cut)
            slice_numbers.append(slice_num)
            particle_counts.append(count)

        except Exception as e:
            print(f"Error processing slice {slice_num}: {e}")
    else:
        print(f"Missing file: {file_path}")

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(slice_numbers, particle_counts, linestyle='-', color='blue')
plt.xlabel('Slice')
plt.ylabel('Particle Count (Y between 50 and 137.25 µm)')
plt.title('Particle Count vs Slice (Central 87.25 µm Only)')
plt.grid(True)

if particle_counts:
    plt.ylim(1, max(particle_counts))

plt.tight_layout()
plt.savefig('particle_count_hardcut_50um.png')
plt.show()
```

**The below particles utilise the same file of data, however particles are counted with different size cuts. I.e. different thresholds include 0.01 μm^2, > 0.05 μm^2, > 0.1 μmm^2 etc. Note that 0.01 μm^2 is the size for one pixel.**

The graph below plots particle count vs slice for particles larger than 0.0001$\mu\mathrm{m}^2$ in size.


```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Directory where CSVs are stored
data_dir = r'C:\Paleo-Detection\output_csv_summary_2-20'
max_slice = 3546

# Define fixed hard Y-bounds (µm)
slice_top = 0
slice_bottom = 187.25
margin = 50

y_lower_cutoff = slice_top + margin          # 50 µm from top = 50
y_upper_cutoff = slice_bottom - margin       # 50 µm from bottom = 137.25

# Minimum particle size threshold in µm²
min_area = 0.0001

slice_numbers = []       # To store slice indicies
particle_counts = []     # To store number of particles passing filters per slice

for slice_num in range(1, max_slice + 1):
    file_path = os.path.join(data_dir, f'results_slice_{slice_num}.csv')
    if os.path.exists(file_path):
        try:
            df = pd.read_csv(file_path, header=0) # Read csv file into dataframe

            # Check necessary columns exist
            if 'Y' not in df.columns:
                print(f"Slice {slice_num} missing 'Y' column.")
                continue
            if 'Area' not in df.columns:
                print(f"Slice {slice_num} missing 'Area' column.")
                continue

            # Convert columns to numeric, drop rows with invalid data
            df['Y'] = pd.to_numeric(df['Y'], errors='coerce')
            df['Area'] = pd.to_numeric(df['Area'], errors='coerce')
            df = df.dropna(subset=['Y', 'Area'])

            # Apply filters: Y position and minimum size
            df_cut = df[(df['Y'] >= y_lower_cutoff) & (df['Y'] <= y_upper_cutoff) & (df['Area'] > min_area)]

            count = len(df_cut) # Count filtered particles in this slice
            slice_numbers.append(slice_num)
            particle_counts.append(count)

        except Exception as e:
            print(f"Error processing slice {slice_num}: {e}")
    #else:
        #print(f"Missing file: {file_path}")

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(slice_numbers, particle_counts, linestyle='-', color='blue')
plt.xlabel('Slice')
plt.ylabel(f'Particle Count (Y between {y_lower_cutoff} and {y_upper_cutoff} µm, Area > {min_area} µm²)')
plt.title(f'Particle Count vs Slice (Central {y_upper_cutoff - y_lower_cutoff} µm, Size > {min_area} µm²)')
plt.grid(True)

if particle_counts:
    plt.ylim(1, max(particle_counts))

plt.tight_layout()
plt.savefig('particle_count_hardcut_sizecut_50um_0.0001um2.png')
plt.show()
```

## `Heatmap of 2D density plot projection in the XY Plane`

Aggregates X and Y positions of all particles across slices, filtering out small particles (Area ≥ 0.0001 µm²) and those within 50 µm of the slice’s top or bottom. The x-axis is left untouched (not cut). The cleaned coordinates are binned into a 2D histogram to produce a heatmap, showing the spatial density distribution of particles in the XY plane. Below, all slices are analysed.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# === Settings ===
data_dir = r'C:\Paleo-Detection\output_csv_summary_2-20'
max_slice = 3546

# Hard Y-bounds
slice_top = 0
slice_bottom = 187.25
margin = 50
y_lower_cutoff = slice_top + margin
y_upper_cutoff = slice_bottom - margin

min_area = 0.0001  # Area threshold in µm²

# Bin settings
num_x_bins = 200
num_y_bins = 100

# Will store all X and Y values across all slices
x_all = []
y_all = []

for slice_num in range(1, max_slice + 1):
    file_path = os.path.join(data_dir, f'results_slice_{slice_num}.csv')
    if not os.path.exists(file_path):
        continue

    try:
        df = pd.read_csv(file_path)

        if 'X' not in df.columns or 'Y' not in df.columns or 'Area' not in df.columns:
            continue

        df['X'] = pd.to_numeric(df['X'], errors='coerce')
        df['Y'] = pd.to_numeric(df['Y'], errors='coerce')
        df['Area'] = pd.to_numeric(df['Area'], errors='coerce')
        df = df.dropna(subset=['X', 'Y', 'Area'])

        # Apply area and Y filtering (cut off top and bottom margin)
        df_cut = df[
            (df['Y'] >= y_lower_cutoff) &
            (df['Y'] <= y_upper_cutoff) &
            (df['Area'] > min_area)
        ]

        x_all.extend(df_cut['X'].tolist())
        y_all.extend(df_cut['Y'].tolist())

    except Exception as e:
        print(f"Error in slice {slice_num}: {e}")

# Create 2D histogram (heatmap)
heatmap, x_edges, y_edges = np.histogram2d(
    x_all,
    y_all,
    bins=[num_x_bins, num_y_bins]
)

# Plotting
plt.figure(figsize=(10, 6))
plt.imshow(
    heatmap.T,  # Transpose to align axes correctly
    cmap='hot',
    interpolation='nearest',
    origin='lower',
    aspect='auto',
    extent=[x_edges[0], x_edges[-1], y_edges[0], y_edges[-1]]
)
plt.colorbar(label='Particle Count')
plt.xlabel('X Position (µm)')
plt.ylabel('Y Position (µm)')
plt.title(f'XY Heatmap of Particles')
plt.tight_layout()
plt.savefig('xy_particle_heatmap.png', dpi=300)
plt.show()
```

## `Uranium Concentration and Sample Age Calculations`

The following code uses the particle density calcualtion to calcualte the uranium concentration, and hence the age of the Mica Bitotite sample. Note that the assumed Uranium concentration of biotie is 1 ppm, but in reality this can be anywhere between 1 ppm - 1 ppb.

```python
import math

def calculate_uranium_and_age(track_density):
    # --- Constants ---
    biotite_density_g_per_um3 = 3e-12  # g/µm³ (from density 3 g/cm³)
    uranium_mass_fraction = 1e-6       # 1 ppm = 1 µg U per g biotite
    uranium_molar_mass = 238.04        # g/mol
    avogadro = 6.023e23                 # atoms/mol
    decay_constant = 0.000155125        # Myr⁻¹ (given)

    # Uranium atoms per gram
    u_atoms_per_g = (uranium_mass_fraction / uranium_molar_mass) * avogadro

    # Uranium atoms per cubic micron
    U_per_um3 = biotite_density_g_per_um3 * u_atoms_per_g

    # Sample age from track density
    t_Myr = math.log(1 + track_density / U_per_um3) / decay_constant

    return U_per_um3, t_Myr


if __name__ == "__main__":
    N = float(input("Enter track density (tracks/µm³): "))
    U_per_um3, age_Myr = calculate_uranium_and_age(N)

    print(f"\nUranium atoms per cubic micron: {U_per_um3:.2e} atoms/µm³")
    print(f"Sample age: {age_Myr:.5f} Myr")
```
